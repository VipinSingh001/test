```python
# Database - shared collection of logically related data
# Advantage -  Data Storage, Data Analysis, Record Keeping, Web Application
# Operation on Database - CRUD 
# Create, Retrive, Update, Delete
# Properties of Ideal Database -  Integrity(Accuracy + consistency), Availability(no down time), Security, Independent of Application, Concurrency(Parallel usage)

# Types of Databases - 

#1.Relational Database(row based database) - SQL databases which store data in the form of row and column(table form)
# mysql,postgres,oracle,sqlserver, microsoft access

#2.NoSQL(Not only Sequel ) Database - handle large amount of unstructured data
# MongoDB

#3.Column Database(fast w.r.t. row) - used in data warehousing and analytical application
# Amazon Redshift, Google BigQuery,showflake
# Example - consider these are rows
# vip 30 B.tech
# akshat 30 B.tech
# data will store in column database like this
# vip akshat 
# 30 30 
# B.tech B.tech

 
# Row oriented databases - are databases that organize data by record, keeping all of the data associated with a record next to each other in memory    
# Example - Postgres, MySQL
# vip 30 B.tech
# akshat 30 B.tech
# data store like--- vip 30 B.tech akshat 30 B.tech




# Graph DataBase - used in recommendation system and Social Network Connection
# Amazon Neptune, Neo4j

# Key-Value database - store data in the form of key-value well suited for caching
# Redis, AmazonDynamoDB 


# Database terminology

# table is known as relation
# columns name - attributes and fields
# row - records or tuples
# if there are 5 columns and 4 rows
# Degree of relation - 4
# cardinality of the relation - 5

# DBMS - provide the interface which needed to store,organise and manage data in database. 
# Users --interact with > Applications --run on> DBMS-- provide gui of database> Database --store data in memory > OS
# function of DBMS - Data management, Integrity, concurrency, Transaction, Security, Utilities


# Database Keys
# A key in database is an attribute or set of attributes that uniquely identifies a row in a table

# Super Key(Janta) - set of colulmns that uniquely identifies ---- email+ name+ rollnumber can be super key

# Candidate Key(umeedwar) - minimal set of super key has no redundant attributes and that is used to uniquely identify a row --- rollnumber,
#email is smallest set and can be candidate key

# primary key(winner from the election) - is candidate key that is uniquely identified and can not be null and dublicate --- like email or rollnumber
# it can not be null and dublicate
#small,numerical,constant -- attributes of primary key
# Alternate key - is also candidate key that is not used as primary key

# candidate key - primary key = alternate key


# composite key - is a primary key that is made up of two or more attributes(more than one primary key is composite key)
# when single key is not sufficient to uniquely identified

# Sarogate key - when table do not have any uniquely indentified attributes(column) and you add one column(id) to make primary key is called sarogate key
# Foreign key - is primary key from one table that is used to establish a relationship with another table

# Cardinality of Relationship
# one to one - person and license number - one table required to store it
# one to many - student and college branch or Restaurants and orders -- two table is required to store it(restaurants and orders two table is required)
# many to many - (Restaurants and menu) or (student and courses) - 3 tables to store the info (student, courses, enrollment)



# Drawbacks - Complexity, Cost, Scalability, Data Integrity, Security, Data Migration, Flexibility
```
